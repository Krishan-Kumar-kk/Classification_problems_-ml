

# TEST AND TRAIN SPLIT



```
from sklearn.model_selection import train_test_split
```

```
X = df.drop('Class', axis = 1).values
y= df['Class'].values
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```
X_train.shape
```

```
y_train
```

# MODEL 

```
#1. DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix # evaluation metric
from sklearn.metrics import accuracy_score # evaluation metric
from sklearn.metrics import f1_score # evaluation metric
```

```
tree_model = DecisionTreeClassifier(max_depth = 4, criterion = 'entropy')
```

```
tree_model.fit(X_train, y_train)
tree_yhat = tree_model.predict(X_test)
```

```
accuracy_score(y_test, tree_yhat)
```

```
f1_score(y_test, tree_yhat)
```

```
#2 KNeighborsClassifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```
n=5
knn = KNeighborsClassifier(n_neighbors = n)
knn.fit(X_train, y_train)
knn_yhat = knn.predict(X_test)
```

```
accuracy_score(y_test, knn_yhat)
```

```
from sklearn.linear_model import LogisticRegression # Logistic regression algorithm
from sklearn.svm import SVC # SVM algorithm
from sklearn.ensemble import RandomForestClassifier # Random forest tree algorithm
from xgboost import XGBClassifier # XGBoost algorithm
```

```
# 3. Logistic Regression

lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_yhat = lr.predict(X_test)
```

```
# 4. SVM 

svm = SVC()
svm.fit(X_train, y_train)
svm_yhat = svm.predict(X_test)
f1_score(y_test,svm_yhat)
```

```
# 5. Random Forest Tree

rf = RandomForestClassifier(max_depth = 4)
rf.fit(X_train, y_train)
rf_yhat = rf.predict(X_test)
```

```
# 6. XGBoost

xgb = XGBClassifier(max_depth = 4)
xgb.fit(X_train, y_train)
xgb_yhat = xgb.predict(X_test)
```

```
print("Compare accuracy score:")
print("accuracy_score of  DecisionTree: ",accuracy_score(y_test, tree_yhat))
print("------------------------------------------------------------------------")
print("accuracy_score of  KNeighborsClassifier: ",accuracy_score(y_test, knn_yhat))
print("------------------------------------------------------------------------")
print("accuracy_score of  Logistic Regression: ",accuracy_score(y_test,lr_yhat))
print("------------------------------------------------------------------------")
print("accuracy_score of  Random Forest Tree: ",accuracy_score(y_test, rf_yhat))
print("------------------------------------------------------------------------")
print("accuracy_score of  XGBoost: ",accuracy_score(y_test, xgb_yhat))
print("------------------------------------------------------------------------")
print("accuracy_score of   SVM: ",accuracy_score(y_test, svm_yhat))
```

```
print("Compare F1 Score:")
print("f1_score of  DecisionTree: ",f1_score(y_test, tree_yhat))
print("------------------------------------------------------------------------")
print("f1_score of  KNeighborsClassifier: ",f1_score(y_test, knn_yhat))
print("------------------------------------------------------------------------")
print("f1_score of  Logistic Regression: ",f1_score(y_test,lr_yhat))
print("------------------------------------------------------------------------")
print("f1_score of  Random Forest Tree: ",f1_score(y_test, rf_yhat))
print("------------------------------------------------------------------------")
print("f1_score of  XGBoost: ",f1_score(y_test, xgb_yhat))
print("------------------------------------------------------------------------")
print("f1_score of   SVM: ",f1_score(y_test, svm_yhat))
```
